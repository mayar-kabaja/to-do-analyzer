```python
import pandas as pd
df = pd.read_csv("../data/cleaned_tasks.csv")
print(df[df["priority"] == "high"]["text"].head(5))
print(df[df["priority"] == "low"]["text"].head(5))
```

## Priority distribution

```python
priority_counts = df["priority"].value_counts()
print(priority_counts)
import matplotlib.pyplot
priority_counts.plot(kind="bar", title="Priority distribution", xlabel="Priority", ylabel="Count");
```

## Category distribution

```python
category_counts = df["category"].value_counts()
print(category_counts)
try:
    import matplotlib.pyplot
    category_counts.plot(kind="bar", title="Category distribution", xlabel="Category", ylabel="Count");
except ImportError:
    print("(Chart skipped: install matplotlib with the first cell, then restart kernel and run again.)")
```
